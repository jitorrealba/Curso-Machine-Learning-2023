# FILTRADO

## Analizamos que productos son los mas relevantes

```python
df_sum = df.groupby(['AÑO',"MES", 'PRODUCTO'])['MONTOFOB'].sum().reset_index()
df_sum
```

```python
df_año = df_sum[df_sum.AÑO == 2017].groupby(['PRODUCTO'])['MONTOFOB'].sum()
df_año.sort_values(ascending=False)
```

```python
df_año = df_sum[df_sum.AÑO == 2018].groupby(['PRODUCTO'])['MONTOFOB'].sum()
df_año.sort_values(ascending=False)
```

```python
df_sum.AÑO.unique()
```

```python
for i in df_sum.AÑO.unique():
    df_año = df_sum[df_sum.AÑO == i].groupby(['PRODUCTO'])['MONTOFOB'].sum()
    df_año = df_año.sort_values(ascending=False)
    print("")
    print(i)
    print("")
    print(df_año)
    
```

## Seleccionamos 'Cobre', 'Minerales de cobre y sus concentrados', 'Conservas y preparaciones de pescados y mariscos', 'Celulosa', dejamos fuera "resto no minería" por ser una categoría de agrupación

## Reemplazamos en la columna MES los numeros correspondientes

```python
replacement_dict = {
        'Enero': "01",
        'Febrero': "02",
        'Marzo': "03",
        'Abril': "04",
        'Mayo': "05",
        'Junio': "06",
        'Julio': "07",
        'Agosto': "08",
        'Septiembre': "09",
        'Octubre': "10",
        'Noviembre': "11",
        'Diciembre': "12"
        }

df_sum['MES'] = df_sum['MES'].replace(replacement_dict)
df_sum
```

Filtrado por productos que nos interesan

```python
df2 = df_sum[df_sum["PRODUCTO"].isin(['Cobre','Minerales de cobre y sus concentrados','Conservas y preparaciones de pescados y mariscos','Celulosa'])]
```

```python
df2.head(5)
```

```python
df2['AÑO']= df2['AÑO'].astype(str)
df2['time_concatenate'] = df2[['MES', 'AÑO']].agg('-'.join, axis=1)
df2
```

```python
df2['month_year'] = pd.to_datetime(df2['time_concatenate'], format='%m-%Y')
df2.drop("time_concatenate", axis = 1, inplace = True)
df2
```

```python
df2 = df2.sort_values('month_year')
df2
```

```python
df2.dtypes
```

```python
df2.PRODUCTO.unique()
```

```python
df2 = df2.drop(["AÑO","MES"], axis = 1)
df2
```

```python
df2 = df2.pivot_table(index='month_year', columns='PRODUCTO', values='MONTOFOB')

# Reset the index of the pivoted dataframe
df2 = df2.reset_index()

# Print the pivoted dataframe
df2
```

```python
df2.set_index('month_year', inplace = True)
```
